# Australia's Participation Rate in Review

Australia's unemployment rate had an unexpected drop from 5.6% to 5.4% for May 2018. I wanted to quickly review the participation rate across the states of Australia.

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline

#import the data source: http://www.abs.gov.au/AUSSTATS/abs@.nsf/DetailsPage/6202.0May%202018?OpenDocument

df = pd.read_excel("6202012.xlsx", sheet_name="Data3", skiprows=range(1,10), parse_dates=[0], mangle_dupe_cols=True)

#Set time as the index
date = df.index.name = 'time'


#Define the Columns I want to use - Participation Rate (Seasonally Adjusted) and rename them


df = df.rename(columns={'Participation rate ;  Persons ;  > New South Wales ;.1': 'NSW',
                       'Participation rate ;  Persons ;  > Victoria ;.1': 'VIC',
                       'Participation rate ;  Persons ;  > Queensland ;.1': 'QLD',
                       'Participation rate ;  Persons ;  > South Australia ;.1': 'SA',
                       'Participation rate ;  Persons ;  > Western Australia ;.1': 'WA',
                       'Participation rate ;  Persons ;  > Tasmania ;.1': 'TAS',
                       'Participation rate ;  Persons ;  > Northern Territory ;.1': 'NT',
                       'Participation rate ;  Persons ;  > Australian Capital Territory ;.1': 'ACT'})


df = df[['NSW', 'VIC','QLD','SA','WA','TAS','NT','ACT']]

#Line Plot to see the difference in participation rates between states

df.tail(48).plot(kind='line', figsize=(35,35), linewidth=7.0)
plt.yticks(size = 40)
plt.xticks(size = 40)
plt.ylabel('Participation (%)', size = 40, labelpad=30)
plt.xlabel('Time', size = 40)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop={'size': 50})
plt.title("Participation Rates Between States - Seasonally Adjusted", size = 60, y=1.02)
```

Across the states participation rates generally look stable/slight increase apart from a steep decline in ACT and slight declines in Vic and Qld over the last few months. A quick review of the media stories appears to not speak about this decline, but has instead noted that ACT has had a strong increase in employment and states a low unemployment rate.

Did ACT/VIC/QLD drop in participation rate improve the national unemployment rate from 5.6% to 5.4%?
